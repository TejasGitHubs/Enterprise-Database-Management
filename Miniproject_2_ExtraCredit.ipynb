# 2. Find events for a given block

```python
# Events are called Events because they are very good at signalling that an event has taken place. 
#In blockchain terms, they’re cheap. They’re cheap to write in a contract in terms of deployment cost and gas cost when calling a function with an event, 
#and they’re free to read. In short, they’re a very good way of broadcasting relevant information.
%%sql
SELECT * FROM event
WHERE event.block_number=7075271
Limit 10
```

# 4. Find specific events for a given contract

### I used the address of the TenXPay token contract again, however we would like to know the values of transfers greater than 1ETH:

```python
# Specific events for given contract will help me to understand the Smart contract return values for the user interface, Asynchronous triggers with data
# and A cheaper form of storage.
%%sql
SELECT arg->'scaled', arg->'num'
FROM "event",jsonb_array_elements(args) arg
WHERE event = 'Transfer' AND address = '0xB97048628DB6B661D4C2aA833e95Dbe1A905B280'
AND (arg->'num')::numeric > 1000000000000000000
LIMIT 10
```

# 5. Find the latest 10 Dai Transfer events and extract sender, receiver and value from JSON

```python
# The latest Dai Transfer will help me understand Staking of idle tokens to produce interest income, Borrowing DAI to buy other coins, 
#Using it as a stable investment to hold profits from other token sales and For in-game purchases.
%%sql
SELECT
      *,
      args->0->>'hex' as "from",
      args->1->>'hex' as "to",
      CAST(args->2->'scaled' AS NUMERIC) AS "value"
FROM event
WHERE address = '0xB97048628DB6B661D4C2aA833e95Dbe1A905B280'
AND event = 'Transfer'
ORDER BY timestamp DESC
LIMIT 10
```
